**4.Write a Python program to read the data randomly of 100 samples and split the data into 80% training data and
display it without using sklearn**

```python
import pandas as pd
import numpy as np

import random

# Replace "your_data" with the actual data you want to split
data =data = pd.read_csv('customer.csv')  # Sample data

# Shuffle the data randomly
#random.shuffle(data)

# Number of samples to read
num_samples = 100
data = data[:num_samples]  # Select the first 100 samples

# Calculate split point for 80/20 split
split_point = int(0.8 * len(data))

# Separate training and test data
training_data = data[:split_point]
test_data = data[split_point:]

# Display training data
print("Training Data:")
for item in training_data:
    print(item)

# Display test data
print("\nTest Data:")
for item in test_data:
    print(item)
```

**10.Write a Python program to read the data randomly of 100 samples and split the data into 20% testing data and
display it without using sklearn**

```python
import pandas as pd
import numpy as np

import random

# Replace "your_data" with the actual data you want to split
data=pd.read_csv('customer.csv')  # Sample data



# Number of samples to read
num_samples = 100
data = data[:num_samples]  # Select the first 100 samples

# Calculate split point for 20/80 split
split_point = int(0.2 * len(data))

# Separate training and test data
training_data = data[split_point:]
test_data = data[:split_point]  # Note the swapped order for 20% test data

# Display training data
print("Training Data:")
for item in training_data:
    print(item)

# Display test data
print("\nTest Data:")
for item in test_data:
    print(item)
```
